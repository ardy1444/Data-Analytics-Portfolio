FAQ BOT

---
This bot is created to replicate a bank chat bot with mutiple functionalities only using python code to create everything from scratch. 
It uses data stored in dictory and iteratively goes through each to fulfill the conditions. 

---
Functions have been defined to check pin, and other credentials and to return desirable output with the user. 



```python
#[First name, Account balance, Account PIN, Last date and time of last transaction, Account Type, Credit Limit]
bank_account_details= {
    1234: [ 'Aaradhya', 200200, 1234, '11-04-2020 22:02:20', 'Platinum', 500000],
    1235: ['Aana', 100000 , 3811, '11-30-2020 13:34:12', 'Basic', 150000],
    1236: ['Meena', 10000, 1234, '11-29-2020 11:11:11', 'Basic', 9999]
}
def verify_support_pin(account_number, support_pin):
    '''This function will verify whether the support pin is correct for the given account number or not. So that our chatbot would be able to decide whether to proceed further or not.'''
    account_details= bank_account_details[account_number]
    pin= account_details[2]
    if (pin== support_pin):
        return True
    else:
        return  False

def check_account(account_number):
    '''This function will return in case if the account number is not in bank. '''
    if (account_number in bank_account_details):
        return True
    else:
        return False
def check_credit_limit(account_number):
    '''This function will return the credit limit remaining in the credit card account. '''
    account_details= bank_account_details[account_number]
    if account_details[1] >= account_details[5]:
        return False
    else:
        return True
def view_balance(account_number):
    '''This function will return account balance of a specific account.'''
    account_details= bank_account_details[account_number]
    return account_details[1]
def platinum_not(account_number):
    '''This function is to verify that a user is a platinum card member or not.'''
    account_details= bank_account_details[account_number]
    if account_details[4] == 'Platinum':
        return True
    else:
        return False
#Name variable is being established to greet the user and account number is to chweck if number is one of the accounts or not..
name= input('Assistant: Hey! Nice to see you! What is your name? \n User: ')
account_number= int(input('Assistant: What is your account number \n'+ name +':'))
while check_account(account_number) == False:
    account_number= int(input('Assistant: Please enter a valid account number.\n'+ name +':'))

pin= int(input('Assistant: What is your pin for the account?\n'+ name +':'))
while verify_support_pin(account_number, pin) == False:
    pin= int(input('Assistant: You have entered the wrong pin. Please enter the right pin.\n'+ name+ ':'))


chat_status= True
depth=0
while chat_status:
    if depth== 0:
        choice= int(input('''Assistant: Thank you for verification. Are you facing any of these issue?
        (1) Credit card getting declined.
        (2) Credit Card showing error.
        (3) You see an unidentified charge.
        (4) To see your balance.
        (5) To apply for a platinum credit card.
        (6) To know more about loan
        (0) To exit \n
'''+ name +':'
        ))
        if choice== 0:
            chat_status= False
        depth= depth+1         
    if depth== 1:
        if choice==1 :
            if check_credit_limit(account_number):
                response= input('Assistant: We have gone through your account details. Your usage seems to be in limit. We have sent your details to our senior executives. He will be calling you on your registered number soon. Please confirm your mobile number.\n'+name+':')
                depth=0
            else:
                response= input('Assistant: Your account has exceeded the credit limit. To deposit money in your card press 1 or return to main menu press 0.')
                if response== '1':
                    print('Assistant: You will soon recieve payement link on your mobile number. Thank you have a nice day.')
                    depth=0
                else:
                    depth=0
        elif choice==2 :
            response = int(input("""Assistant: Have you tried following?
            (1)Do you remember blocking your card?
            (2)Swiping your card with an international carrier
            """ + name + ':'))
            while True:

                if response== 1:
                    print('Assistant: Please call 1800-200-5924 for further help.')
                    depth=0
                    break
                elif response ==2:
                    print('Your card will be unblocked in next 24 hours.')
                    depth=0
                    break
                else:
                    response = int(input("""Assistant: Please select a valid option. 
                    (1)Do you remember blocking your card?
                    (2)Swiping your card with an international carrier
              """ +name + ':'))
        elif choice== 3:
            print('Please call 1800-200-5924 for further help.')
            depth=0
        elif choice== 4:
            print('Assistant: Your account balance is $'+ str(view_balance(account_number)))
            depth=0
        elif choice== 5:
            if platinum_not(account_number)== True:
                print('Assistant: You already have a platinum credit card. ')
                depth=0
            else:
                print('Assistant: Our costumer executive will contact you in next 10 minutes.')
                depth=0
        elif choice== 6:
            response= input('Assistant:Please enter your phone number at which you will be reachable at. \n'
                            +name+':')
            print('Assistant: Our constumer executive will contact you in next 5 minutes to brief you about your loan.')
            depth=0
```
